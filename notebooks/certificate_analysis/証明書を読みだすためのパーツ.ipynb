```python
import pandas as pd
from sqlalchemy import create_engine

# データベース接続設定
DB_CONFIG = {
    "host": "localhost",
    "dbname": "website_data",
    "user": "postgres",
    "password": "asomura"
}

def get_engine(db_name: str):
    return create_engine(f"postgresql://{DB_CONFIG['user']}:{DB_CONFIG['password']}@{DB_CONFIG['host']}/{db_name}")

# データ取得
def fetch_data(db_name: str, table_name: str):
    engine = get_engine(db_name)
    query = f"SELECT domain, domain_registrar, whois_date, https_certificate_issuer, https_certificate_expiry, hosting_provider, ip_address FROM {table_name} WHERE status = 7;"
    return pd.read_sql_query(query, engine)

website_df = fetch_data("website_data", "website_data")
normal_df = fetch_data("normal_sites", "normal_sites")
```
