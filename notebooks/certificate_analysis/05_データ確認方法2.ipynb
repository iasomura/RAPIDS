```python
# 分析結果確認プログラム（簡略版）
from pathlib import Path
import json
from datetime import datetime

def print_analysis_results():
    """Print key analysis results and file information"""
    # パス設定
    base_dir = Path('/home/asomura/waseda/nextstep/RAPIDS')
    data_dir = base_dir / 'data' / 'processed'
    output_dir = base_dir / 'reports' / 'certificate_analysis'
    log_dir = base_dir / 'data' / 'logs'

    # JSONファイルの読み込みと表示
    json_files = sorted(data_dir.glob('cert_features_*.json'), 
                       key=lambda x: x.stat().st_mtime, reverse=True)[:2]
    
    results = {}
    for file_path in json_files:
        with open(file_path) as f:
            db_name = 'website_data' if 'website_data' in file_path.name else 'normal_sites'
            results[db_name] = json.load(f)
            print(f"\nResults from {file_path.name}:")
            print(json.dumps(results[db_name], indent=2))

    # 画像ファイル一覧
    print("\nGenerated visualization files:")
    for img_path in sorted(output_dir.glob('*.png')):
        file_time = datetime.fromtimestamp(img_path.stat().st_mtime).strftime('%Y-%m-%d %H:%M:%S')
        print(f"- {img_path.name} (Created: {file_time})")

    # ログ表示
    latest_log = max(log_dir.glob('advanced_cert_analysis_*.log'), 
                    key=lambda x: x.stat().st_mtime)
    with open(latest_log) as f:
        print(f"\nLog content ({latest_log.name}):")
        print(f.read())

if __name__ == "__main__":
    print_analysis_results()
```
